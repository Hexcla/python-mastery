Why Multi Threading

What is Threading?

- A thread in computer science refers to a separate flow of execution within a process.
- A process is an instance of a program that is being executed, and each process can have one or more threads.
- Threads within the same process share the same memory space and resources, allowing them to communicate and coordinate with each other more efficiently compared to separate processes.

 key points to understand about threads

- Concurrent Execution
- Lightweight
- Shared Resources
- Multithreading: Multithreading refers to the use of multiple threads within a single process to perform concurrent tasks.
- Thread Safety: When multiple threads access shared resources concurrently, it's essential to ensure thread safety to avoid data corruption and synchronization issues.

Example

```python
import threading
import time


def print_numbers():
    for i in range(5):
        print(i)
        time.sleep(1)  # Simulate some time-consuming task

def print_letters():
    for letter in 'ABCDE':
        print(letter)
        time.sleep(10)  # Simulate some time-consuming task

# Create two threads
thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_letters)

# Start the threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

print("Both threads have finished.")
```

```python
import threading

def student_task(task_name):
    print(f"{task_name} is solving the math problem")

# Creating threads for each student
student1 = threading.Thread(target=student_task, args=("Student 1",))
student2 = threading.Thread(target=student_task, args=("Student 2",))

# Starting the threads (students begin working)
student1.start()
student2.start()

# Waiting for each student to finish their task
student1.join()
student2.join()

print("All students have finished solving the math problems. Moving on to the next activity.")
```

Synchronization

```python
import threading
import time

# Shared resource
pencil = "Pencil"

# Mutex for synchronization
mutex = threading.Lock()

# Function to write with the pencil
def write_on_blackboard(student_name):
    global pencil
    # Acquire the lock before accessing the shared resource
    mutex.acquire()
    print(f"{student_name} is writing on the blackboard with {pencil}")
    # Simulate writing time
    time.sleep(1)
    print(f"{student_name} finished writing")
    # Release the lock after accessing the shared resource
    mutex.release()

# Create student threads
students = []
for i in range(5):
    student_name = f"Student {i+1}"
    student_thread = threading.Thread(target=write_on_blackboard, args=(student_name,))
    students.append(student_thread)
    student_thread.start()

# Wait for all students to finish
for student_thread in students:
    student_thread.join()

print("All students have finished writing on the blackboard.")
```
