```python
# gussing game

import random

target = random.randint(1,100)
#print(target)
guess = 0

while guess != target:
    guess = int(input("enter the number between 1 to 100: "))
    #print(target)

    
    if guess > target:
        print("try with lower number")
    elif guess < target:
        print("try with larger number")
    else:
        print("You guessed correctly")
```

```python
import random

target = random.randint(1,100)
#guess = 0
attempt = 0


while True:
    guess = int(input("enter the number between 1 to 100: "))
   # print(target)
    attempt += 1
    
    if guess > target:
        print("try with lower number")
    elif guess < target:
        print("try with larger number")
    else:
        print("You guessed correctly")
        print("Total Attempt Taken :- ", attempt)
        break
```
