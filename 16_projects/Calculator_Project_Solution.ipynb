# Calculator Project By Using Python Function

```python
# Function for Addition
def add(x, y):   #user-defined fuction
    return x + y
```

```python
# Function for Subtraction
def subtract(x, y):
    return x - y
```

```python
# Function for Multiplication
def multiply(x, y):
    return x * y
```

```python
# Function for Division
def divide(x, y):
    if y == 0:
        raise ValueError("Cannot divide by zero.")
    return x / y
```

```python
# Function for Modulus
def modulus(x, y):
    if y == 0:
        raise ValueError("Cannot perform modulus with zero.")
    return x % y
```

# Scientific Calculations:

```python
import math

# Function for Exponentiation
def exponentiation(x, y):
    return x ** y
```

```python
# Function for Square Root
def square_root(x):
    if x < 0:
        raise ValueError("Cannot calculate square root of a negative number.")
    return math.sqrt(x)
```

```python
# Function for Trigonometric Calculations
def sine(x):
    return math.sin(math.radians(x))

def cosine(x):
    return math.cos(math.radians(x))

def tangent(x):
    return math.tan(math.radians(x))
```

```python
# Function for Logarithms
def logarithm(x, base):
    if x <= 0 or base <= 0 or base == 1:
        raise ValueError("Invalid input for logarithm calculation.")
    return math.log(x, base)
```

```python
# Function for Factorials
def factorial(x):
    if x < 0:
        raise ValueError("Cannot calculate factorial of a negative number.")
    if x == 0:
        return 1
    return x * factorial(x - 1)
```

```python
# Main Calculator Function
def calculator():
    print("Python Calculator")
    print("Select operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Modulus")
    print("6. Exponentiation")
    print("7. Square Root")
    print("8. Sine")
    print("9. Cosine")
    print("10. Tangent")
    print("11. Logarithm")
    print("12. Factorial")

    while True:
        try:
            choice = int(input("Enter choice (1-12): "))

            if choice < 1 or choice > 12:
                print("Invalid choice. Please try again.")
                continue

            if choice in [7, 8, 9, 11, 12]:
                x = float(input("Enter number: "))
            else:
                x = float(input("Enter first number: "))
                y = float(input("Enter second number: "))

            if choice == 1:
                print("Result:", add(x, y))
            elif choice == 2:
                print("Result:", subtract(x, y))
            elif choice == 3:
                print("Result:", multiply(x, y))
            elif choice == 4:
                print("Result:", divide(x, y))
            elif choice == 5:
                print("Result:", modulus(x, y))
            elif choice == 6:
                print("Result:", exponentiation(x, y))
            elif choice == 7:
                print("Result:", square_root(x))
            elif choice == 8:
                print("Result:", sine(x))
            elif choice == 9:
                print("Result:", cosine(x))
            elif choice == 10:
                print("Result:", tangent(x))
            elif choice == 11:
                base = float(input("Enter base: "))
                print("Result:", logarithm(x, base))
            elif choice == 12:
                print("Result:", factorial(x))

            break

        except ValueError as e:
            print("Error:", e)

if __name__ == "__main__":
    calculator()
```
