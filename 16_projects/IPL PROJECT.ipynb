# 6. filter out the match data which were conducted in city Bangalore and store the same data into the file.

```python
specific_value = 'Bangalore'
bangalore_data = df[df['city'] == specific_value]
bangalore_data.to_csv('bangalore_data.csv', index=False)
```

```python
df_bang = pd.read_csv("bangalore_data.csv")
df_bang
```

# 7. Now display the data which is stored in our file. consist of matches which were played excluding the city Bangalore.

```python
specific_value = 'Bangalore'
exclude_bang = df[df['city'] != specific_value]

# Store the filtered data into a new CSV file
exclude_bang.to_csv('exclude_bang.csv', index=False)
```

```python
df_exc_bang = pd.read_csv("exclude_bang.csv")
df_exc_bang.head(4)
```
