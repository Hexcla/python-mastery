# 1. count plot

```python
#Create a count plot of the "class" column, grouped by "who"

#ax = sns.countplot(x="class",hue="who", data = titanic)
ax = sns.countplot(x=titanic["class"],hue=titanic["who"])
plt.show()
```

# 2.Scatter plot

```python
a = [1,2,3,4,5]
b = [3,5,4,6,8]

sns.scatterplot(x=a,y=b)

#add labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Display the plot
plt.show()
```

# 3.line plot

```python
a = [1,2,3,4,5]
b = [3,5,4,6,8]

# Create a line plot using Seaborn
sns.lineplot(x=a, y=b, color = "red")

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')

# Display the plot
plt.show()
```

# 4.Bar Plot

```python
# Sample data
a = ['A', 'B', 'C', 'D', 'E']
b = [3, 5, 4, 6, 8]

# Create a bar plot using Seaborn
sns.barplot(x=titanic["class"],y=titanic["survived"])

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot Example')
#plt.grid(axis="y") # both

# Display the plot
plt.show()
```
