# Numpy--Arithmetic Operations (OR) Matrix Operations

```python

a = np.array([[10,20],[30,40]])
b = np.array([[1,2],[3,4]])

d = np.add(a,b)

print(d)
```

```python
d.ndim
```

```python
d = d.reshape(4)

print(d,d.ndim)
```

```python
a=np.array([[10,20],[30,40]])
b=np.array([[1,2],[3,4]])

#print(a)
#print(b)

c = a + b

print(c,type(c))

```
